Linear Regression Implementation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, r2_score


def linear_reg(X_train, y_train, X_test, y_test):

    model = LinearRegression()
    clf = model.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    linear_reg_score = r2_score(y_test, y_pred)
    coeffieicent = clf.coef_ #testing in coefficients 
    
    return linear_reg_score, coeffieicent
```

```python
linreg_score_list = [] 
pipeline_score_list = []
logisReg_score_list = []
coefficietnts_list = [] 

for i in range(len(targets)):
  print(targets[i])
  
  y = original_df[targets[i]]  # Loop through all targets

  y = y.replace(np.nan, 0)

  df_to_visualize = pd.concat([X, y], axis=1)

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, train_size = 0.66, random_state=42)
  
  acc_score, coeffieicent = linear_reg(X_train, y_train, X_test, y_test)
  

  print(acc_score)
  print(coeffieicent.mean())
  
  linreg_score_list.append(acc_score)
  coefficietnts_list.append(coeffieicent.mean())

target_score = list(zip(targets, linreg_score_list, coefficietnts_list))
score_df = pd.DataFrame(target_score, columns=['Targets', 'Linear_Reg_Score','coefficient avg'])
```
