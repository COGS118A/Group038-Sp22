### Multiple Linear Regression (Subsection 1)
We will be predicting 1 target with 1191 predictors. We assume that the target is a linear function of its predictors, and thus model the relationship by a high-dimensional line with linear regression. We will use LinearRegression() method to create a linear regression object and got our independent and dependent training values in as parameters into our regression model. Once we have our regression model fitted, we can predict using our test set. We will return coefficients to see the relationship. <br>

Our finding (Multiple Linear Regression):
1. First getting all the coefficients of the training_features (predictors) and taking the mean to see if on average of our training_targets would increase or decrease. The **result** we got was **negative** meaning that our training_targets decreases by that many factor on average.
2. We also look at the Mean Square Error (as our Metric) to see if our prediction was good/bad compared to the expected. If our score is close to 0, that implies our predictions are good. The **result** that we got are around **0.05 - 0.01** which implies that our predictions are very good compared to the expected values. 

```python
# Create empty lists to append later
linreg_score_list = []
coef_list = []

def linear_reg(X_train, y_train, X_test, y_test):

    # Define model
    model = LinearRegression()
    # Fit our training data to our model
    clf = model.fit(X_train, y_train)
    # Predict on test set
    y_pred = clf.predict(X_test)

    # Getting MSE
    linear_reg_score = mean_squared_error(y_test, y_pred)
    # Getting coefficients
    coefficient = clf.coef_ 

    return linear_reg_score, coefficient

for i in range(len(targets)):
  
  y = era_1_df[targets[i]]  # Loop through all targets

  X_train, X_test, y_train, y_test = train_test_split(X_1, y, test_size=0.33, train_size = 0.66, random_state=42)
  acc_score, coefficient = linear_reg(X_train, y_train, X_test, y_test)
  
  linreg_score_list.append(acc_score)
  coef_list.append(coefficient.mean())

target_score = list(zip(targets, linreg_score_list, coef_list))
linReg_score_df = pd.DataFrame(target_score, columns=['Targets', 'Linear_Reg_Score','coefficient avg'])

linReg_score_df
```

#### The base model
We will now build the model described in the Proposed Solution section. To recapitulate, a regression is trained, testing data is input batch-style, the percentiles are computed, and the points are classified by which percentiles they fall between. This works because we know the exact distribution of our target.

```python
reg = sm.OLS(y_1.get('target'), X_1).fit()
era_2_reg_predictions = reg.predict(X_2.reset_index().drop(columns='index'))
era_2_reg_predictions
```

```python
era_2_percentiles = [np.percentile(era_2_reg_predictions, i) for i in [5,25,75,95,100]]
def classify_reg_prediction(reg_prediction, percentiles=era_2_percentiles):
    for i in range(5):
        if reg_prediction<=percentiles[i]:
            return i/4

era_2_class_predictions = era_2_reg_predictions.apply(classify_reg_prediction)
print('The accuracy of this model on era 2 data (unseen) is', str(accuracy(y_2.reset_index().get('target'), era_2_class_predictions)))
```

This is worse than chance! (50%)

We now compute the **one-vs-rest accuracy** metric mentioned in the Evaluation Metrics section. Note that this metric is for testing how well a model would have performed if our problem was a 2-class problem, so the accuracies given will be better than the accuracy of the model on the actual problem (2 class accuracy >= 5 class accuracy).

```python
def one_vs_rest_accuracy(actual, predicted):
    for i in [0, 0.25, 0.50, 0.75, 1]:
        class_accuracy = ((predicted==i) == (actual==i)).sum()/predicted.shape[0]
        print('The one-vs-rest accuracy for class', i, 'was', class_accuracy)
        
one_vs_rest_accuracy(y_2.reset_index().get('target'), era_2_class_predictions)
```

These results make it more clear why our model is performing so poorly. The accuracy is great for the smaller classes, but bad for the large class. The large class takes up 50% of the dataset, so if we want to optimize for accuracy, we should pay more attention to it.
