### Numerai Information
Numerai is a global AI tournament, where participants compete against each other to build the best model that predicts the stock market using Machine Learning methods. Numerai provides the competitors with an encrypted dataset, initially a historical data set, and the competitors task is to build a model to predict the future target using live features, corresponding to the current stock market.[7] The users do this by training their data and making predictions on live weekly tournament data.[7]. This tournament runs on a weekly basis and it incentivizes users to compete through large monetary prizes. Numerai then uses all its submissions and winners model to create a “Meta Model”, which rules the capital of the Numerai hedge fund[7]. 

# Problem Statement

Given numer.ai’s dataset of individual stocks characterized by numerical features, we want to predict a target feature measuring performance. Our problem is using current data to find the best model (within our abilities) to use for such a goal. “Best” will be operationalized by performance and generalizability, both of which can be evaluated numerically. Performance will be computed by a metric of prediction power: if we train our model on previously unseen stocks, how well will the targets be predicted? Such a metric will be defined as we explore the data. Since we are working with targets that represent performance, we will perhaps use a loss/reward function that does not penalize cases where the model predicts a high value for the target but the actual value is even higher. Or, we will not penalize these cases as much as those where the model predicts a high target value but the actual target value was low (or vice versa). That is, we will build a loss/reward function that cares more about predicting the direction of the target (positive or negative, high or low, above average or below average, etc.) than it cares about predicting the magnitude (very above average vs. extremely above average, etc.) of the target. Generalizability will be evaluated by “variance”: that means, the variance of the expected value of the generalization error. This is a metric commonly used to estimate generalizability, and is often the “variance” referred to in the “bias/variance tradeoff”. From this numerical measure we can make qualitative conclusions about our model’s propensity to overfit in the context of generalizability. These metrics will be computed by following scientific guidelines like randomization and partitioning (into train-validation-test). We will compute these metrics in a notebook that is runnable by graders.

##### Perform Linear Regression

```python
linreg_score_list = []
coef_list = []

def linear_reg(X_train, y_train, X_test, y_test):

    # Define model
    model = LinearRegression()
    # Fit our training data to our model
    clf = model.fit(X_train, y_train)
    # Predict on test set
    y_pred = clf.predict(X_test)
    # Getting r^2 score
    linear_reg_score = r2_score(y_test, y_pred)
    coefficient = clf.coef_ #testing in coefficients 
    
    return linear_reg_score, coefficient

for i in range(len(targets)):
  
  y = original_df[targets[i]]  # Loop through all targets

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, train_size = 0.66, random_state=42)
  
  acc_score, coefficient = linear_reg(X_train, y_train, X_test, y_test)
  
  linreg_score_list.append(acc_score)
  coef_list.append(coefficient.mean())

target_score = list(zip(targets, linreg_score_list, coef_list))
score_df = pd.DataFrame(target_score, columns=['Targets', 'Linear_Reg_Score','coefficient avg'])

score_df
```

##### Decision Tree Implementation
After seeing negative scores in Linear Regression, we were unsure if our data really fit  into Linear Regression Form. (Ingrid) decided to implement Decision Tree to look at the top (say ~20) important features (predictors) in our dataset. 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

def decisionTreeRegress(X_train, X_test,y_train, y_test): 
    regr1 = DecisionTreeRegressor(max_depth = 4, random_state=0)
    clf = regr1.fit(X_train, y_train)
    regr1_pred = regr1.predict(X_test)
    
    #Perform 5-fold cross validation 
    crossScores = cross_val_score(estimator=clf, X=X, y=y, cv=5)

    #Compute r^2 score 
    scores = clf.score(X_test, y_test)
    
    #getting parameters 
    getParams = clf.get_params

    #pull out top 20 important features, descending order
    for importance, name in sorted(zip(clf.feature_importances_, X_train.columns),reverse=True)[:20]:
         print(name, importance)

    return crossScores, scores, getParams

crossValScoreList = [] 
actualScoresList = []
for i in range(len(targets)):
  
  y = original_df[targets[i]]  

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, train_size = 0.66, random_state=42)

  crossValScore, acc_score, paramDetails = decisionTreeRegress(X_train, X_test, y_train, y_test)

#   print(crossValScore)
  
  #avg cross val score 
  crossValScoreList.append(crossValScore.mean())

  #r^2 score for each i 
  actualScoresList.append(acc_score)
  
crossValZip = list(zip(targets,crossValScoreList, actualScoresList))
ValScore_df = pd.DataFrame(crossValZip, columns=['Targets', 'CrossValScoresMean', 'Scores'])

ValScore_df
    
```

##### KNN Classifier
After seeing so many negative score, we started to wonder if our data really fit  into our models. Hence, (Thanh) decided to try something different and fit the data into a classifier model to see if something interesting might pop up. 

```python
def knn_classifier_score(X_train, X_test, y_train, y_test):
    knn = KNeighborsClassifier(n_neighbors=200, weights='distance', metric='euclidean', n_jobs=-1)
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    acc_score = accuracy_score(y_test, y_pred)
    
    return acc_score

knn_score_list = []
for i in range(len(targets)):
    y = original_df[targets[i]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    lab_enc = preprocessing.LabelEncoder()
    y_train_encoded = lab_enc.fit_transform(y_train)  # Encoded cuz getting 'continuous' variable error
    y_test_encoded = lab_enc.fit_transform(y_test)    # Encoded cuz getting 'continuous' variable error
    
    knn_score = knn_classifier_score(X_train, X_test, y_train_encoded, y_test_encoded)
    
    knn_score_list.append(knn_score)
    knn_target_score = list(zip(targets, knn_score_list))
    knn_score_df = pd.DataFrame(knn_target_score, columns=['Targets', 'KNN_Accuracy_Score'])
    
knn_score_df
    
```
