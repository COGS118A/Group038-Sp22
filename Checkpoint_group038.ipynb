### Numerai Information
Numerai is a global AI tournament, where participants compete against each other to build the best model that predicts the stock market using Machine Learning methods. Numerai provides the competitors with an encrypted dataset, initially a historical data set, and the competitors task is to build a model to predict the future target using live features, corresponding to the current stock market.[7] The users do this by training their data and making predictions on live weekly tournament data.[7]. # Problem Statement

Given numer.ai’s dataset of individual stocks characterized by numerical features, we want to predict a target feature measuring performance. Our problem is using current data to find the best model (within our abilities) to use for such a goal. “Best” will be operationalized by performance and generalizability, both of which can be evaluated numerically. Performance will be computed by a metric of prediction power: if we train our model on previously unseen stocks, how well will the targets be predicted? Such a metric will be defined as we explore the data. Since we are working with targets that represent performance, we will perhaps use a loss/reward function that does not penalize cases where the model predicts a high value for the target but the actual value is even higher. Or, we will not penalize these cases as much as those where the model predicts a high target value but the actual target value was low (or vice versa). That is, we will build a loss/reward function that cares more about predicting the direction of the target (positive or negative, high or low, above average or below average, etc.) than it cares about predicting the magnitude (very above average vs. extremely above average, etc.) of the target. Generalizability will be evaluated by “variance”: that means, the variance of the expected value of the generalization error. This is a metric commonly used to estimate generalizability, and is often the “variance” referred to in the “bias/variance tradeoff”. From this numerical measure we can make qualitative conclusions about our model’s propensity to overfit in the context of generalizability. These metrics will be computed by following scientific guidelines like randomization and partitioning (into train-validation-test). We will compute these metrics in a notebook that is runnable by graders.

### Multiple Linear Regression
We will be predicting 1 target with 1191 predictors. We assume that the target is a linear function of its predictors, and thus model the relationship by a high-dimensional line with linear regression. We will use LinearRegression() method to create a linear regression object and got our independent and dependent training values in as parameters into our regression model. Once we have our regression model fitted, we can predict using our test set. We will return coefficients to see the relationship. <br>

Our finding (Multiple Linear Regression):
1. First getting all the coefficients of the training_features (predictors) and taking the mean to see if on average of our training_targets would increase or decrease. The **result** we got was **negative** meaning that our training_targets decreases by that many factor on average.
2. We also look at the Mean Square Error (as our Metric) to see if our prediction was good/bad compared to the expected. If our score is close to 0, that implies our predictions are good. The **result** that we got are around **0.05 - 0.01** which implies that our predictions are very good compared to the expected values. 


```python
# Create empty lists to append later
linreg_score_list = []
coef_list = []

def linear_reg(X_train, y_train, X_test, y_test):

    # Define model
    model = LinearRegression()
    # Fit our training data to our model
    clf = model.fit(X_train, y_train)
    # Predict on test set
    y_pred = clf.predict(X_test)

    # Getting MSE
    linear_reg_score = mean_squared_error(y_test, y_pred)
    # Getting coefficients
    coefficient = clf.coef_ 

    return linear_reg_score, coefficient

for i in range(len(targets)):
  
  y = era_1_df[targets[i]]  # Loop through all targets

  X_train, X_test, y_train, y_test = train_test_split(X_1, y, test_size=0.33, train_size = 0.66, random_state=42)
  acc_score, coefficient = linear_reg(X_train, y_train, X_test, y_test)
  
  linreg_score_list.append(acc_score)
  coef_list.append(coefficient.mean())

target_score = list(zip(targets, linreg_score_list, coef_list))
linReg_score_df = pd.DataFrame(target_score, columns=['Targets', 'Linear_Reg_Score','coefficient avg'])

linReg_score_df
```

### Decision Tree Implementation
After seeing positive scores in Linear Regression, indicating a sort of correlation we wanted to try to understand what are the most important and relevent features that predict our data. As we have over 1000 predictors, we can use a decision tree regressor for each target, to figure out if any predictors stand out. We took out the top 20, in descending order, are results are as follows. We will calculate our accuracy, using cross validation with a 5 k fold. We will be comparing this to the r^2 score and see if there are anything that pops out. 

```python
# Method to build decision Tree Regress Model 
def decisionTreeRegress(X_train, X_test,y_train, y_test): 
    regr1 = DecisionTreeRegressor(max_depth = 4, random_state=0)
    clf = regr1.fit(X_train, y_train)
    regr1_pred = regr1.predict(X_test)
    
    #Perform 5-fold cross validation 
    crossScores = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=5)

    #Compute r^2 score 
    scores = clf.score(X_test, y_test)
    
    #getting parameters 
    getParams = clf.get_params

    #pull out top 20 important features, descending order
    for importance, name in sorted(zip(clf.feature_importances_, X_train.columns),reverse=True)[:20]:
         print(name, importance)

    return crossScores, scores, getParams

# Create empty lists to append later
crossValScoreList = [] 
actualScoresList = []

for i in range(len(targets)):
  
  y = era_1_df[targets[i]]  

  X_train, X_test, y_train, y_test = train_test_split(X_1, y, test_size=0.33, train_size = 0.66, random_state=42)

  crossValScore, acc_score, paramDetails = decisionTreeRegress(X_train, X_test, y_train, y_test)
  
  #avg cross val score 
  crossValScoreList.append(crossValScore.mean())

  #r^2 score for each i 
  actualScoresList.append(acc_score)
  
crossValZip = list(zip(targets,crossValScoreList, actualScoresList))
ValScore_df = pd.DataFrame(crossValZip, columns=['Targets', 'CrossValScoresMean', 'Scores'])

ValScore_df
    
```

As shown above, we can see that each target, has around **14-18 important predictors** that make up the decision tree. **This means that, out of the 1191 features, only a tiny amount really matter!** 

Following that, we have our cross validation scores, for each target along with the **r^2 score**, which indicate some **negative correlation**, where the model is indicating it isnt fitting as best as it could. <br>

When implementing our decision tree, I tried a varition of max depths from about 4-8, and I continued to see negative scores.
